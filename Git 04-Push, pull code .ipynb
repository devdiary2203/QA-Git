# Git 04: Push/push code 

Sau khi tạo xong một project trên gitlab, bạn cần clone project về để có thể code và push code từ máy cá nhân (local repository) lên gitlab (remote repository). Để thực hiện các thao tác này, ta cần biết cách sử dụng 2 lệnh ```git remote``` và ```git push```

## Lệnh: git remote
Lệnh này giúp git có thể theo dõi project trên gitlab và kết nối với project trên máy bạn (local). Git tạo một tên tham chiếu đại diện cho project lưu trên máy bạn kết hợp với link project trên gitlab để tạo liên kết giữa máy chủ từ xa (remote) và máy cá nhân (local). Tên tham chiếu này hiểu đơn giản là một cái tên để người dùng có thể dễ hiểu, dễ thao tác hơn khi thực hiện các lệnh push, pull code từ máy chủ remote thay vì phải copy một đoạn link dài của project trên trình duyệt.

Đoạn này có thể hơi khó hiểu, mình sẽ minh họa rõ hơn phía dưới:

```git remote add origin https://gitlab.com:2317/devdiary2203/test-project.git```

Lệnh trên tạo một liên kết giữa local và project trên gitlab, liên kết này được đặt tên là *origin*. Tên này có thể thay đổi được theo cú pháp:

```git remote add name_remote link_project_on_gitlab_or_github```

Đối với một project được tạo, khi clone về bạn không cần tạo remote này nữa, tên remote mặc định thường dùng là *origin*. Để kiếm tra tên của remote, bạn có thể dùng lệnh ```git remote```:

=> Tên remote được đặt trên project là origin

## Lệnh git push
Ở phần trước, sau khi code và update cho git về các thay đổi bằng lệnh ```git add``` và ```git commit```, bạn cần update các thay đổi này lên remote project ở gitlab bằng lệnh ```git push```.


Cú pháp lệnh này như sau:
```git push <remote_repository> <branch_name>```

*remote_repository* chính là tên remote mà chúng ta đặt ở trên, mặc định được đặt là origin

Ví dụ muốn push các thay đổi code lên branch main:
```git push origin main```



Như hình trên, mình đã push code của project này lên github (gitlab tương tự)

## Lệnh git fetch/git merge
Trong quá trình làm một project có nhiều thành viên tham gia, có thể có những thay đổi, cập nhật mới trên project của thành viên khác mà bạn không biết, ví dụ như:
- Thêm, xóa, sửa file
- Tạo/xóa các branch
Để cập nhật những thay đổi này lệnh ```git fetch``` được sử dụng để download các thay đổi từ máy chủ remote về máy của bạn.
Cú pháp lệnh này như sau:

```git fetch origin```

*origin* là tên remote, nếu không truyền thì git sẽ mặc định sử dụng tên remote đang có.




Sau khi download các thay đổi mới nhất về máy, nếu bạn thấy các thay đổi này ok thì bạn có thể merge các thay đổi này vào một branch nào đó mà bạn muốn, ví dụ muốn merge các thay đổi vào branch main như sau:

```git merge origin/main```

## Lệnh git pull

Tương tự như ```git fetch```, ```git pull``` sẽ download các thay đổi mới nhất từ máy chủ remote (github, gitlab,...) trên một branch nào đó nhưng sẽ thêm bước merge luôn vào project local của bạn chứ không cần dùng lệnh ```git merge```.

Cú pháp của lệnh này: 

```git pull origin branch_name```

Ví dụ muốn download các thay đổi từ branch main:

```git pull origin main```



Thông báo *Already up to date.* nghĩa là code của bạn đang ở trạng thái mới nhất so với code trên remote project ở gitlab.

Lệnh này sẽ hữu ích khi bạn muốn cập nhật những thay đổi mới nhất từ một branch nào đó và để tránh tình trạng conflict code trước khi push code lên gitlab.

```git add``` -> ```git commit``` -> ```git pull``` (cập nhật thay đổi) -> ```git push```
