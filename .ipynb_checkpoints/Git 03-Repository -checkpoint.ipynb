# Git 03: Repository
Github, gitlab,... là các dịch vụ lưu trữ và quản lí source code sử dụng git là công cụ để thao tác. Bạn có thể tạo repository trên các dịch vụ này để lưu trữ và quản lí project của mình.

Một số thành phần cơ bản của project khi khởi tạo:
- README.md: File này thường được tạo để bạn note các thông tin chính của project như mục đích project, hướng dẫn cách chạy, mô tả các tính năng, các lỗi có thể xảy ra. Đối với nhiều project được public thì còn có thêm thông tin bản quyền, contact tác giả... File này được viết dưới dạng ngôn ngữ markdown: https://www.markdownguide.org/cheat-sheet
- .gitignore: Tên gọi của nó đã nói lên ý nghĩa. Có một vài tình huống:
    + Khi code bạn có thể thao tác trên các hệ điều hành hoặc công cụ code khác nhau, việc này có thể sinh ra các file không cần thiết đối với project của bạn như file .vscode khi sử dụng công cụ Visual Studio Code, file .DSStore trên macos hay file pycache khi trong quá trình chạy code python
    + Bạn tạo một file mysql_connect_test.py chỉ để kiểm tra connection từ máy của bạn đến một database mysql và không muốn commit file này vào project.
    
    Các trường hợp trên có thể được giải quyết bằng cách bạn thêm đường dẫn file hoặc thư mục mà bạn không muốn commit vào file .gitignore. Sau này, khi thực hiện commit, git sẽ kiểm tra trong file .gitignore để bỏ qua các file hoặc thư mục mà bạn đã thêm.

## Git clone
Để copy một project từ các dịch vụ như github, gitlab về máy cá nhân để code và commit lại thì ta sẽ sử dụng lệnh ```git clone``` theo cú pháp:

```git clone link_gitlab_project```

Copy link project tại đây:


Bạn có thể chọn link dạng https hoặc ssh để clone project của mình về máy. Trong trường hợp chỉ muốn clone code trên một branch nào đó, bạn có thể thêm option *-b* + tên branch để clone:

```git clone -b branch_name link_gitlab_project```

Ví dụ:
```git clone -b dev link_gitlab_project```
