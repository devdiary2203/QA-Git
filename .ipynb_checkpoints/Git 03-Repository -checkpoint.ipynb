## Git clone
 copy mt project t cc dch v nh github, gitlab v my c nhn  code v commit li th ta s s dng lnh ```git clone``` theo c php:

```git clone link_gitlab_project```

Copy link project ti y:


Bn c th chn link dng https hoc ssh  clone project ca mnh v my. Trong trng hp ch mun clone code trn mt branch no , bn c th thm option *-b* + tn branch  clone:

```git clone -b branch_name link_gitlab_project```

V d:
```git clone -b dev link_gitlab_project```
