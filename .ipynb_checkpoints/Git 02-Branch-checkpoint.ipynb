# Git 02: Branch
Một vài lợi ích lớn mà git mang lại là giúp cho việc phát triển project trong môi trường teamwork hiệu quả hơn cũng như hoàn thiện project nhanh hơn bằng cách triển khai song song các tính năng. 

Có một tình huống rất hay gặp phải: Trong một project, khi bạn đang phát triển tính năng A nhưng sếp lại có thêm yêu cầu cho bạn phát triển gấp một tính năng B trước để ra mắt cho khách hàng. Lúc này nếu thực hiện luôn trên chính code mà bạn đang sửa cho tính năng A rất dễ gây ra nhầm lẫn và gặp lỗi. Đây chính là lí do vì sao cần dùng đến branch trong git.

Khi khởi tạo project, branch chính mặc định được đặt tên là master hoặc main tùy theo dịch vụ bản sử dụng. Tên branch chính này có thể thay đổi khi bạn tạo commit đầu tiên. Bạn có thể tạo các branch khác từ một branch gốc để phát triển các yêu cầu riêng. Lúc này các commit code tiếp theo của bạn trên các branch sẽ tách biệt với nhau, sẽ được merge lại vào branch chính khi tạo merge request. Một logic tạo các branch thường thấy: 

main -> dev 

        |
            
        dev -> feature-tinh_nang_A
            
            -> feature-tinh_nang_B
            
            -> fix-loi_C

## Lệnh ```git branch```
Bây giờ ta sẽ tìm hiểu cách dùng lệnh git branch.

#### Tạo một branch mới: 
Ta có thể tạo một branch mới từ branch hiện tại bằng cách dùng cú pháp: ```git branch your_branch_name```. Đối với các branch tạo ra nhằm mục đích code tính năng hay fix lỗi nên có các title phía trước chú thích như "feature-", "fix-" để dễ quản lí và review (tương tự như title khi commit message). Ví dụ:

```git branch dev```

```git branch feature-tinh_nang_A```

#### Liệt kê các branch đang có trên máy của bạn:

```git branch```



Như trên hình, mình vừa tạo branch dev từ branch main. Sau đó kiểm tra lại thì đang có 2 branch trên máy mình:
- main
- dev

Dấu * trước branch main cho biết hiện tại bạn đang làm việc trên branch main. Nếu bạn muốn chuyển qua branch dev để làm việc thì sử dụng lệnh ``` git checkout branch_name``` như sau:

    ```git checkout dev```



Mình vừa chuyển qua làm việc trên branch dev. Sau đó kiểm tra lại bằng lệnh ```git branch``` thì thấy dấu * đang trỏ về branch dev.

Bạn có thể chuyển không gian làm việc qua lại giữa branch theo cú pháp ```git checkout branch_name```. Trước khi chuyển qua branch khác cần commit các thay đổi trên branch hiện tại trước nhé.

Một cách khác kết hợp vừa tạo branch mới và chuyển sang làm việc trên branch đó luôn bằng cách sử dụng lệnh ```git checkout -b branch_name```. Lưu ý phải có option **-b** ở đây.

Ví dụ trên mình tạo branch mới feature-tinh_nang_A từ branch dev.

#### Đổi tên branch
Khi đang code tính năng thống kê chỉ số A nhưng khách hàng lại thích gọi đó là chỉ số B nên bạn muốn đổi tên branch để dễ quan sát hơn. Có 2 cách đổi tên một branch:
- Cách 1: Checkout sang branch và đổi tên theo cú pháp như sau:

    ```git checkout branch_name```
    
    ```git branch -m new_branch_name```

- Cách 2: Đổi tên một branch bất kì theo cú pháp:

    ```git branch -m old_name new_name```

Branch tên feature-tinh_nang_A giờ đã được đổi tên thành feature-tinh_nang_B

#### Xóa branch
Khi đã hoàn thành xong một tính năng và merge code thành công vào branch gốc, nếu không cần sử dụng nữa thì bạn có thể xóa branch tính năng này đi để project trông đỡ rối hơn bằng cú pháp ```git branch -d branch_name```.
Lưu ý phải có option **-d** và trước khi xóa thì bạn phải checkout sang một branch khác trước.

Ví dụ mình đã hoàn thành xong tính năng B, giờ mình sẽ xóa branch ```feature_tinh_nang_B``` đi:

Như trong hình, mình checkout sang branch dev trước sau đó xóa branch ```feature_tinh_nang_B``` đi. Kiểm tra lại thì project của chúng ta chỉ còn 2 branch là ```main``` và ```dev```

#### git stash
Ở trên mình có nói trước khi chuyển qua branch khác cần commit các thay đổi trên branch hiện tại trước nhé. Giả sử bạn đang code cùng lúc nhiều tính năng. Khi đang code tính năng A thì phải chuyển sang fix lỗi cho tính năng B, và không muốn commit các thay đổi hiện tại trên branch của tính năng A do chưa code xong. Lúc này lệnh ```git stash``` có thể bạn sẽ cần. 

```git stash``` sẽ tạm thời lưu trữ các file thay đổi nhưng chưa được commit ở branch hiện tại trong bộ nhớ cache. Sau đó bạn có thể checkout sang branch khác để làm việc tiếp mà không cần phải commit trên branch hiện tại trước. 

Ví dụ bạn đang sửa một file test.txt trên branch feature-tinh_nang_A và muốn chuyển sang branch dev để làm tiếp. Lệnh ```git stash``` sẽ sao lưu tạm thời các thay đổi hiện tại trên branch feature-tinh_nang_A để bạn chuyển sang branch dev làm việc tiếp

Trên đây là một số ví dụ sử dụng lệnh ```git branch``` và một số lệnh liên quan để thao tác với branch trên git. Bài sau mình sẽ mô tả các ví dụ để thao tác giữa code trên máy của bạn với các dịch vụ lưu trữ sử dụng git như github, gitlab.
