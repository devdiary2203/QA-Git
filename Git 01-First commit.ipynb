### Lệnh ```git config```
```git config``` có thể hiểu là một cách để bạn đăng nhập vào git sử dụng username và email mà bạn đăng kí, thường xuất hiện để yêu cầu bạn đăng kí thông tin khi clone một project về máy lần đầu. Lệnh này giúp git quản lí được ai đang đóng góp vào project. ```git config``` có các level để lựa chọn cấu hình:
- locally: ở level này thông tin email, user name bạn đăng kí chỉ có tác dụng trên project hiện tại. Khi clone một project mới về thì các thông tin này không có tác dụng ảnh hưởng lên project mới

```git config --local user.email "devdiary@gmail.com"```

```git config --local user.name "devdiary"```



- globally: ở level này, thông tin user.email, user.name của bạn sẽ có tác dụng trên tất cả các project trên máy, ngoại trừ các project bạn đã đăng kí với thông tin khác từ trước


```git config --global user.email "devdiary@gmail.com"```

```git config --global user.name "devdiary"```

- system: Đây là level cao nhất, thông tin email và name bạn truyền vào trong lệnh git config sẽ áp dụng cho toàn bộ user trên server hoặc trên máy. Thông thường thì không cần dùng đến level này. Đối với máy tính cá nhân thì chỉ cần dùng global level là đủ

### Lệnh ```git add```
Khi bạn code, các file trong project được sửa, xóa hoặc thêm mới. Lúc này file đó vẫn đang nằm trong vùng *Working Area* trên máy bạn và Git chưa cập nhật các thay đổi này để theo dõi. Lệnh ```git add``` sẽ cập nhật cho git các thay đổi trong project. Có nhiều lựa chọn với lệnh ```git add```:
- Thêm một file cụ thể để git theo dõi
- Thêm nhiều file để git theo dõi
- Thêm các file trong một thư mục cụ thể để git theo dõi
- Chỉ thêm các file mà bạn sửa nội dung (bao gồm các file mà bạn tạo mới kiểu new folder, new file trên máy)

Các option này sẽ mô tả chi tiết hơn bên dưới, nhưng trước hết hãy thử qua lệnh ```git status``` của git như sau:

Lệnh ```git status``` này cho phép xem trạng thái các file trong thư mục project mà bạn đang làm việc. Ví dụ trên mình thực hiện trên chính project này. Hiện tại ta đang ở branch main của project. Có một số file được thay đổi nhưng git chưa cập nhật để theo dõi:
- .ipynb_checkpoints/
- Git 01-config, commit, more.ipynb
- Q&A about Git.ipynb

Bây giờ ta sẽ dùng lệnh ```git add```, sau đó kiểm tra lại trạng thái của project thông qua lệnh ```git status```

Sau khi sử dụng ```git add .``` và kiểm tra lại thì git thông báo đã cập nhật các file mới được tạo:
- new file:   .ipynb_checkpoints/Git 01-config, commit, more-checkpoint.ipynb
- new file:   .ipynb_checkpoints/Q&A about Git-checkpoint.ipynb
- new file:   Git 01-config, commit, more.ipynb
- new file:   Q&A about Git.ipynb
Lúc này các file vừa mới thêm đang ở vùng *Staged Area* và sẵn sàng để đến bước commit các thay đổi này vào repository trên máy bạn.

Trước khi nói đến việc commit, ta sẽ nói về các option của lệnh ```git add``` cung cấp mà mình nói đến ở trên để hiểu tại sao phải dùng các option đó:
- ```git add .```: thêm tất cả các file (sửa, xóa, thêm mới) trong project vào git
- ```git add -u```: chỉ thêm các file đã có từ trước mà bạn sửa (không bao gồm các file thêm mới) trong project vào git
- ```git add new_file.txt```: thêm một file cụ thể vào git

Thông thường, để tránh thêm phải các file không cần thiết được sinh ra (có thể do file cache của ngôn ngữ, các file lưu trạng thái của trình soạn thảo code mà bạn sử dụng) mà bạn không kiểm soát được thì có thể sử dụng ```git add -u``` kết hợp ```git add new_file.txt``` để thêm các file cụ thể mà bạn muốn git cập nhật

### Lệnh ```git commit```
Sau khi đã thông báo cho git về các thay đổi mà bạn muốn cập nhật bằng lệnh ```git add```, bạn cần tạo một bản sao lưu để lưu lại các cập nhật này bằng lệnh ```git commit```, sau đó code của bạn sẽ sẵn sàng để cập nhật lên gitlab.

*Lưu ý: Ở đây mình nói đơn giản qua về chức năng của lệnh ```git commit``` thực hiện. Bạn có thể sử dụng mà không gặp vấn đề gì khi mới sử dụng git ở một số lệnh cơ bản. Để nói kĩ hơn về vấn đề này thì mình sẽ đề cập ở một bài khác liên quan đến việc quản lí con trỏ, đọc nhật kí của git.*

Cú pháp của lệnh ```git commit```:
- Dạng cơ bản:

```git commit -m "your message"```
- Dạng chi tiết:

```git commit -m "title commit" -m "your message"```

*your message* chính là chú thích cho những thay đổi mà bạn vừa thực hiện trên code của mình. Ví dụ, khi vừa thêm tính năng thống kê A vào project, commit của bạn có thể viết như sau:

```git commit -m "feat: add feature stats A"```

Message của bạn cần tóm tắt ngắn gọn các thay đổi mà bạn vừa cập nhật. Commit cũng nên được thực hiện ngay khi bạn thực hiện xong một tính năng hay một thay đổi lớn nào đó để thuận tiện hơn cho quá trình review-merge code sau này. Các message nên có thêm title phía trước để dễ phân biệt các thay đổi của bạn hay đồng nghiệp, như ở commit trên mình dùng title là *feat* rồi mới đến message chính để đánh dấu đây là một commit về tính năng. Một số các title bạn có thể dùng như:
- feat: Tính năng
- fix: Fix lỗi 
- docs: Bổ sung tài liệu
- refactor: Refactor code
- test: Test một đoạn code hay tính năng nào đó
- chores: Các thay đổi nhỏ không làm ảnh hưởng đến project khi bạn chỉ một test một file riêng để kiểm tra connect đến spark chẳng hạn
- revert: Đưa project quay trở lại commit trước (cái này mình sẽ nói thêm ở phần sau)
- init: Khởi tạo các file code thực hiện tính năng nào đó lần đầu
